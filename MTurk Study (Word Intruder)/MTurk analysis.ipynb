```python
import csv
import numpy as np

out = []

with open('StudyResults_topic-model_Carina&Greta-Copy1.txt') as csv_file:
    reader=csv.reader(csv_file,delimiter=',')
    for row in reader:
        out.append(row)
        #print(row, '\n')


correct_answers = []
for i in range(1,51):
    correct = out[i][5]
    correct_answers.append(correct)
#print(correct_answers)
#print(len(correct_answers))


correct_counts = []
correct = []

for i in range(1,51): # indexing through correct intruders
    print(i)
    response_lst = []
    for j in range(0,50): # indexing through participants
        myindex = i + 50*j
        resp = out[myindex][19]
        response_lst.append(resp)
    print(response_lst, '\n')  
    resp_correct = correct_answers[i-1]
    x = response_lst.count(resp_correct)
    print('response', x, '\n\n')
    correct.append(x)

print('Correct counts: \n', correct, '\n')

print('In percent: \n')
correct_props = np.array(correct)/50 * 100
print(correct_props)

random_choice = 1/6 * 100
print('Random choice: ',random_choice)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_style("whitegrid")

lst = [i for i in range(1,51)]
plt.figure()
plt.scatter(lst, correct_props, color = 'cadetblue', label = 'Response')
plt.ylabel('Percentage correct response')
plt.xlabel('Trial number')
plt.title('Inference of topic coherence by MTurkers (n=50)')
plt.axhline(y=random_choice, linestyle=':', label = 'Random chance = 16.67%')
plt.legend()

plt.savefig('InferenceOfTopicCoherence_MTurk.jpeg', dpi=300, bbox_inches='tight')
```

```python
print(correct_props)
print(random_choice)

random_choice_vec = np.full((1, 50), random_choice)[0]

from scipy import stats
ttest = stats.ttest_1samp(correct_props,random_choice)
print(ttest)
```

```python
from sklearn.linear_model import LinearRegression
lst = np.array([i for i in range(1,51)])
x = lst.reshape((-1, 1))
y = correct_props

model = LinearRegression().fit(x, y)

r_sq = model.score(x, y)
print('coefficient of determination, R^2:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```
