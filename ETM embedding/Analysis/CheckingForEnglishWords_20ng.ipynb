```python
import enchant
import pickle
vocab = '/om/user/ckauf/ETM-master/data/min_df_2/vocab.pkl'

with open(vocab, 'rb') as f:
    myvocab = pickle.load(f)
    
print("Vocabulary length: ", len(myvocab))

d = enchant.Dict("en_US")

ans = []

English_words = []
nonEnglish_words = []
for word in myvocab:
    ans.append(d.check(word))
    if d.check(word) == True:
        English_words.append(word)
    else:
        nonEnglish_words.append(word)
#print(ans)
print("Number of English words found in vocabulary: ", ans.count(True))

print('**********************************************************************************************')
print('English words')
print(English_words)

print('**********************************************************************************************')
print('Non-English words')
print(nonEnglish_words)

# Proportion of non-English words 0,489
```
